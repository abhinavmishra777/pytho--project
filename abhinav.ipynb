```python
# Electric Vehicle Data Analysis Project
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

```python

# Task 1: Filter and Analyze Dataset
def task_1(dataset_path):
    df = pd.read_excel(dataset_path)
    filtered = df[(df['Minimal price (gross) [PLN]'] <= 350000) & (df['Range (WLTP) [km]'] >= 400)]
    grouped = filtered.groupby('Make')
    avg_battery_capacity = grouped['Battery capacity [kWh]'].mean()
    return filtered, grouped, avg_battery_capacity

# Task 2: Outliers Detection
def task_2(df):
    column = 'mean - Energy consumption [kWh/100 km]'
    q1, q3 = np.percentile(df[column], [25, 75])
    iqr = q3 - q1
    lower_bound, upper_bound = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Task 3: Visualization of Battery Capacity vs Range
def task_3(df):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Battery capacity [kWh]', y='Range (WLTP) [km]', data=df)
    plt.title('Battery Capacity vs Range')
    plt.xlabel('Battery Capacity (kWh)')
    plt.ylabel('Range (WLTP) (km)')
    plt.show()

# Task 4: Recommendation System
class EVRecommender:
    def _init_(self, df):
        self.df = df

    def recommend(self, budget, desired_range, min_battery_capacity):
        recommended = self.df[(self.df['Minimal price (gross) [PLN]'] <= budget) &
                              (self.df['Range (WLTP) [km]'] >= desired_range) &
                              (self.df['Battery capacity [kWh]'] >= min_battery_capacity)]
        return recommended.nlargest(3, 'Range (WLTP) [km]')

# Task 5: Hypothesis Testing
def task_5(df):
    tesla = df[df['Make'] == 'Tesla']['Engine power [KM]']
    audi = df[df['Make'] == 'Audi']['Engine power [KM]']
    t_stat, p_value = ttest_ind(tesla, audi)
    return t_stat, p_value

# Driver Code
'''if _name_ == "_main_":
    dataset_path = "/content/FEV-data-Excel.xlsx"

    # Task 1
    filtered, grouped, avg_battery_capacity = task_1(dataset_path)
    print("Filtered EVs:", filtered)
    print("Average Battery Capacity by Manufacturer:", avg_battery_capacity)

    # Task 2
    outliers = task_2(filtered)
    print("Energy Consumption Outliers:", outliers)

    # Task 3
    task_3(filtered)

    # Task 4
    recommender = EVRecommender(filtered)
    recommendations = recommender.recommend(350000, 400, 50)
    print("Top Recommendations:", recommendations)

    # Task 5
    t_stat, p_value = task_5(filtered)
    print(f"T-Statistic: {t_stat}, P-Value: {p_value}")

    # Video Explanation Placeholder
    print("Refer to the video link in the accompanying Jupyter Notebook.")'''
```
